```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
# Import the API key.
from config import g_key
```

```python
# Store the CSV you saved created in part one into a DataFrame.
hotel_df = pd.read_csv("weather_data/WeatherPy_vacation.csv")
hotel_df.head(20)
```

```python
#Create separate DataFrames for each city on the travel route
clarksville_df = hotel_df.loc[hotel_df["City"] == "Clarksville"]
saint_anthony_df = hotel_df.loc[hotel_df["City"] == "Saint Anthony"]
sturgeon_bay_df = hotel_df.loc[hotel_df["City"] == "Sturgeon Bay"]
evanston_df = hotel_df.loc[hotel_df["City"] == "Evanston"]
```

```python
clarksvilleTup = ()
clarksvilleTup = clarksville_df.to_numpy()[0][5:7]
```

```python
evanstonTup = ()
evanstonTup = evanston_df.to_numpy()[0][5:7]
```

```python
sturgeon_bayTup = ()
sturgeon_bayTup = sturgeon_bay_df.to_numpy()[0][5:7]
```

```python
saint_anthonyTup = () 
saint_anthonyTup = saint_anthony_df.to_numpy()[0][5:7]
```

```python
import gmaps
import gmaps.datasets
gmaps.configure(api_key=g_key)

# Latitude-longitude pairs
clarksville = clarksvilleTup
evanston = evanstonTup
sturgeon_bay = sturgeon_bayTup
saint_anthony = saint_anthonyTup

fig = gmaps.figure()
vacay_iten = gmaps.directions_layer(clarksville, saint_anthony, waypoints=[evanston, sturgeon_bay], travel_mode='BICYCLING')
fig.add_layer(vacay_iten)
fig
```

```python
# Concatenate DFs
frames = [clarksville_df, saint_anthony_df, sturgeon_bay_df, evanston_df]
vacay_itenDF = pd.concat(frames).reset_index()
vacay_itenDF
```

```python
# Add a pop-up marker for each city.
locations = vacay_itenDF[["Lat", "Lng"]]
fig = gmaps.figure(center=(41.5, -90.4), zoom_level=4)
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description}, {Max Temp}°F</dd>
</dl>
"""
# Store the DataFrame Row.
iten_info = [info_box_template.format(**row) for index, row in vacay_itenDF.iterrows()]

marker_layer = gmaps.marker_layer(locations, info_box_content=iten_info)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```
